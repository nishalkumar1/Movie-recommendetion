```python
import csv
import numpy as np
import pandas as pd
a= pd.read_csv('/content/movies.csv')
b= pd.read_csv('/content/ratings.csv')


```

```python
a.head()
```

```python
b.head()
```

```python
c=pd.merge(a,b,on='movieId')
c.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
%matplotlib inline
```

```python
c.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
c.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
r=pd.DataFrame(c.groupby('title')['rating'].mean())
r.head()
```

```python
r['no of rat']= pd.DataFrame(c.groupby('title')['rating'].count())
r.head()
```

```python
mov= c.pivot_table(index='userId', columns='title', values='rating')
mov.head()
```

```python
r .sort_values('no of rat',ascending=False).head(20)
```

```python
xx=input('enter movie name')
```

```python
xx_ratings= mov[xx]
xx_ratings.head()
```

```python
similar_to_xx= mov.corrwith(xx_ratings)
```

```python
corr_xx= pd.DataFrame(similar_to_xx, columns=['Correlation'])
corr_xx.dropna(inplace=True)
corr_xx.head() 
```

```python
corr_xx.sort_values('Correlation',ascending=False).head()
```

**MOVIES WITH MAX CORRELATION WILL BE REC**OMMENDED
